## Landfast Sea Ice Exent
### What is Landfast Sea Ice?

The landfast sea ice data in this dataset are the same data analyzed by [Mahoney et al. (2007)](https://agupubs.onlinelibrary.wiley.com/doi/abs/10.1029/2006JC003559). Landfast ice is sea ice that is mostly stationary and attached to land. This dataset defines landfast ice as being continguous to the coast and lacking detectablel motion for approximately 20 days. This defintion is consistent with the defintion provided by Mahoney et al. (2005) and Mahoney et al. (2007).


### Spatial Attributes
The spatial domain of this dataset spans northern Alaska and northwestern Canada and includes the Beaufort and Chukchi Seas. Some small islands are excluded from the analysis. Note that the dataset analyzed in Mahoney et al. (2007) may contain a set of western Chukchi data that is not included in the NSIDC holdings. The native spatial resolution (i.e., pixel size) of these data is 100 m x 100 m.

### Temporal Atrributes
Eight annnual (1996-2004) seasonal ice cycles (October to July) are included in these data.

### Origin
The landfast ice data are derived from RADARSAT-1 data aquired using the the moderate-resolution ScanSAR operation mode of the synthetic aperture radar (SAR) sensor. The  RADARSAT-1 mission (no longer active) used a C-Band (5.3 GHz), 5.6 cm wavelength, HH polarized microwave frequency to image the earth. RADARSAT-1 could "detect" landfast ice because it has a distinct backscatter signature. Indiviudal ScarSAR passes were mosaiced together to encompass the study area, with the entire study area being covered about once every 10 days. Data processing details are provided by [Eicken et al. (2006)](https://nsidc.org/sites/nsidc.org/files/files/data/noaa/g02173/eicken_leads_landfast_2006.pdf).


### Why Care About Landfast Sea Ice?
From Mahoney et al. (2007):

>In the Arctic, landfast sea ice is a key element of the coastal system, integral to a wide range of geological and biological processes as well as human activities. The presence of landfast ice can mitigate the effect of winter storms on the coast but also impede navigation in the spring. As well as being of great importance to native subsistence activities [Nelson, 1969; George et al., 2004], the presence or absence of landfast in northern Alaska and its stability are of considerable economic importance for offshore development.

### Expected Landfast Ice Flavors
Note that the [User Guide](https://nsidc.org/data/g02173?qt-data_set_tabs=3#qt-data_set_tabs) for this dataset refers to the landfast ice variable in a few different ways. The term "Landfast Ice Extent" is used for the binarized GeoTIFFs of where `1` values (white) indicate landfast ice and `0` values (black) indicate land, open ocean, or non-landfast ice. The term "seaward looking landfast ice edge" (SLIE) is used to indicate vectorized outlines of the landfast ice. SLIE is used almost exclusively by Mahoney et al. (2007) to analyze landfast ice and to compute additional datasets of landfast ice width (LFIW) and water depth at the SLIE.

SLIE data are available from this dataset via annual ice season (e.g., 1997-1997) geodatabase `.mbd` files that are compressed into `.zip` files and are located in the `landfast1996-2004/annual_geodb` directory.

The landfast ice extent rasters are avilable via GeoTIFFs in the `landfast1996-2004/data/binarized_geotiffs` directory. Each `.tif` should have an accompanying `.tfw` file that includes important geographic information. Landast ice extent is also summarized by monthy mean, meadian, maximum, and minimum, are available in gridded or vecotirzed format. These are in the `landfast1996- 2004/monthly_averages` directory.


Note that the source data (the original RADARSAT-1 SAR imagery) is also available from this dataset, but these data are outside the scope of this intial EDA.


## Leads
### What are Leads?
Per Eicken et al. (2006) leads are defined as "any fracture or passage-way through sea ice which is navigable by surface vessels" and the definition is expanded by the work of Lindsay and Rothrock (2005) to include areas of thin ice that are 0.15 m to 2 m thick. In this dataset no distinction is made between open water leads and thin ice areas.

### Spatial Attributes
Extent is slightly larger than the landfast ice extent variable. Spatial resolution is 1.1 km.

### Temporal Atrributes
Eleven annnual (1993-2004) ice cycles are encompassed by these data. Unlike landfast ice extent, the data sequence is irregular at times due to cloud coverage. December to June encompass the majority of the data aquisitions.

### Origin
Leads data were derived from approximately 385 Advanced Very High Resolution Radiometer (AVHRR) images. Unlike SAR, the AVHRR sensor is sensitive to atmospheric conditions and sunlight. Both visible spectrum and infrared (IR) images were used to delineate leads from land and sea ice. For the period of these data, AVHRR instruments were aboard the  following satellites: NOAA-12, NOAA-14, NOAA-15, NOAA-16, and NOAA-17. The radiometer wavelengths (e.g, aboard NOAA-15) include six channels visible (0.58-0.68 µm), near-IR (0.725-1.0 µm), near-IR (1.58-1.64 µm), IR (3.55-3.93 µm), IR (10.3-11.3 µm) and a si(11.5-12.5 µm).[9]

### Why Care About Leads?
Recurring leads in the Beaufort and Chukchi Seas signal the deformation of the ice pack and impact the ocean-atmosphere heat exchange and polar ecosystems (Eicken et al. 2006).

The annual subdirectories within the `data` directory seem to contain the two primary types of files. First there are files with the the `mos` suffix which should refer to mosaics of the source SAR backscatter data. These mosaics also drive the time schema: a file name of `r97_363_365mos.tif` is of a `YY_DDD_DDD` format where the day is given in the day of the year. Each file has a range of days because it encompasses information from several satellite passes which were then mosaiced together. Note that there are many auxiliary files in this directory: the usual shapefile ones, but also some raster pyramids (`.rrd`) and browse `.jpg` images. Also notice the `.e00` format. The `.e00` format is an old-school format for proprietary ESRI ArcInfo Interchange files. They were used to transfer data (think floppy disk era) but are basically ASCII grids.

````python
# set up a few functions to parse filenames
def get_doy(fp):
    try:
        doy = re.match(r'.*([0-3][0-9][0-9]-[0-3][0-9][0-9])', fp).group(1)
    except:
        doy = re.match(r'.*([0-3][0-9][0-9]_[0-3][0-9][0-9])', fp).group(1)
        doy = doy.replace("_", "-")
    return doy

def get_re_year(fp):
    year = re.match(r'.*([1-3][0-9]{3})', fp).group(1)
    return year


def get_year(fp):
    year = fp[1:5]
    return year
````

````python
def generate_binary_raster_di(rstrs):
    
    di = {}
    
    for fp in rstrs:
        trimmed_fp = fp.name[:-4]
        year = get_year(trimmed_fp)
        doy = get_doy(trimmed_fp)
        fp_key = year + "-" + doy
        di[fp_key] = {}
        with rasterio.open(fp) as src:
            arr = src.read(1).astype(float)
            arr[arr == src.nodata] = np.nan
            di[fp_key]["fp"] = fp
            di[fp_key]["arr"] = arr
            di[fp_key]["min"] = np.round(np.nanmin(arr), 2)
            di[fp_key]["max"] = np.round(np.nanmax(arr), 2)
            di[fp_key]["width"] = arr.shape[1]
            di[fp_key]["height"] = arr.shape[0]
            di[fp_key]["year"] = year
            di[fp_key]["pixel_size_m"] = src.profile["transform"][0]
            di[fp_key]["crs"] = src.crs
    return di
````

````python
binary_slie = generate_binary_raster_di(random.sample(slie_fps, 9))
````

````python
def plot_binary_maps(di):

    cmap = plt.get_cmap('gray', 2)
    
    labels = np.array(["Land, Ocean, or Non-landfast Ice","Landfast Ice"])
    
    fig, axs = plt.subplots(3, 3, figsize=(12, 7), sharex=True, sharey=True)
    fig.patch.set_facecolor("white")
    plt.suptitle("Binary SLIE", y=0.9)
    plt.subplots_adjust(hspace=0)
    vmax = 1
    vmin = 0

    
    for ax, k in zip(axs.flatten(), di.keys()):
        ax.imshow(di[k]["arr"], vmin=vmin, vmax=vmax, cmap=cmap, interpolation="none")
        ax.set_title(k)

    cbar_ax = fig.add_axes([0.15, 0.05, 0.7, 0.06])
    fig.colorbar(axs[0,0].images[0], cax=cbar_ax, ticks=[0, 1], orientation="horizontal")
    cbar_ax.set_xticklabels(labels)

plot_binary_maps(binary_slie)
plt.savefig("nsidc_binarized_slie.png", bbox_inches="tight", dpi=144)
````
